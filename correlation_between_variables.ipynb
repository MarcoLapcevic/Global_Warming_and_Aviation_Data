```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_rows
```

## Co2 Emissions

```python
co2_emissions_dataset = pd.read_csv('CO2_Emission.csv')
co2_emissions_dataset = co2_emissions_dataset.iloc[-44:]

plt.figure(figsize=(6, 4))
plt.xlabel("Time (Years)", fontname='Times New Roman',
           fontsize=12)
plt.ylabel("Annual CO₂ emissions (in million metric tons)", fontname='Times New Roman',
           fontsize=12)
plt.title("CO₂ emissions per year", fontname='Monospace',
          fontsize=18)

plt.plot(co2_emissions_dataset['year'], co2_emissions_dataset['co2'])
```

## Turbulance Reports

```python
turbulence_dataset = pd.read_csv('global_warming_turbulence_data.csv')
turbulence_dataset = turbulence_dataset.iloc[-45:]

plt.figure(figsize=(6, 4))
plt.xlabel("Time (Years)", fontname='Times New Roman',
           fontsize=12)
plt.ylabel("Reported Turbulance Events", fontname='Times New Roman',
           fontsize=12)
plt.title("Reported Turbulance events per year", fontname='Monospace',
          fontsize=18)

plt.plot(turbulence_dataset['Year'], turbulence_dataset['Reported_Turbulence_Events'])
```
